## Methodology
### 1. Collect data from Kaggle 
- https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020
- Formula 1 World Championship (1950 - 2024) F1 race data from 1950 to 2024 
- created by Vopani
### 2. Cleaning the data & Feature Engineering
- Load 5 CSVs: results, drivers, races, qualifying, status
- Filter races to 2000–2024 only
- Attach a dnf flag such that anything with status other than "Finished" is a DNF
- Convert Q1/Q2/Q3 lap time strings (e.g. "1:26.572") to seconds, take the best per session, then average per driver
- Aggregate all results into one row per driver
- Compute win_rate, podium_rate, and dnf_rate per driver
- Join qualifying pace back onto the driver table
- Join driver name and nationality for labeling
- Drop drivers with fewer than 30 races in the era
### 3. Dimensionality Reduction (PCA)
- Load data.csv and split it into metadata (names, nationality) and the 8 numeric feature columns
- Impute the 6 missing avg_fastest_lap_speed values using column median
- Standardize all features to mean=0, std=1 so no feature dominates PCA just by having a bigger scale
- Fit PCA on the scaled matrix and return all components and the transformed driver coordinates
- Find minimum number of PCs needed to hit 90% cumulative explained variance
- Extract the loadings table (how much each original feature contributes to each PC)
- Plot a scree plot
    - bar chart of variance explained per PC
- Plot cumulative variance
    - line chart with a 90% threshold line marked
- Plot a biplot
    - drivers scattered in PC space with feature loading arrows overlaid, with a few notable drivers labeled
